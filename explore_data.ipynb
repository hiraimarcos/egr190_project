```python
import torch.nn as nn

class TEST_CNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = nn.Conv1d(in_channels=300, out_channels=80, kernel_size=5, padding=2)
        self.maxpool = nn.MaxPool1d(kernel_size=5)
        self.relu = nn.ReLU()
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(20*80, 100)
        self.fc2 = nn.Linear(100, 1)
        
    def forward(self,x):
        size = len(x)
        x = self.conv(x)
        x = self.relu(x)
        x = self.maxpool(x)
        
        x = self.flatten(x)
        
        x = self.fc1(x)
        x = self.relu(x)
        
        x = self.fc2(x)
        return x.view(size)
        

        
```

```python
model = TEST_CNN()
model
```

```python
model.forward(texts.float())
```

```python
def test_accuracy(model, val_set, num_of_tests):
    ## Testing
    correct = 0
    total = num_of_tests
    val_loader = torch.utils.data.DataLoader(val_set, batch_size=num_of_tests, shuffle=True)
    
    with torch.no_grad():
        # validate on smaller sample so my computer doesn't crash
        data, labels = iter(val_loader).next()
        y = model(data.float())
        predictions = torch.argmax(y.view(-1,1), dim=1)
        correct += torch.sum((predictions == labels).float())
    acc = correct/total
    print('Test accuracy: {}'.format(acc))
    return acc
```

```python
validation = TweeterData_v0('val')
train = TweeterData_v0('train')
```

```python
batch_size = 1000
num_epochs = 3
learning_rate = 0.001
num_iterations = 20

# cross entropy loss didn't work because we are dealing with binary classification
criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate, momentum=0.9)

train_loader = torch.utils.data.DataLoader(train, batch_size=batch_size, shuffle=True)
train_iter = iter(train_loader)

loss_history = {}
loss_history['train'] = []
loss_history['accuracy'] = []

model = TEST_CNN()
model


for iteration in range(num_iterations):
    texts, labels = train_iter.next()
    optimizer.zero_grad()
    # Forward pass
    x = texts
    y = model(x.float())
    loss = criterion(y, labels.float())

    loss_history['train'].append(loss.item())
    loss_history['accuracy'].append(test_accuracy(model, validation, 100))

    # backward pass
    loss.backward()
    optimizer.step()
    print("Loss: {}".format(loss.item()))
    print("{}/{}".format(iteration, num_iterations))
        
```
