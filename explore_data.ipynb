```python
import json, time
import pandas as pd
```

```python
with open('tweets.jsonl', 'r') as json_lines:
    tweet = json_lines.readline()
    tweet = json.loads(tweet)
    print('Name: {}'.format(tweet['user']['name']))
    print('Username: {}'.format(tweet['user']['screen_name']))
    print('Text: {}'.format(tweet['full_text']))
    print('\n')
    print(tweet)
    print('\n')
    print(tweet['user'])
```

```python
unames = set()
users = []

with open('tweets.jsonl', 'r') as json_lines:
    while True:
        tweet = json_lines.readline()
        if tweet == "":
            break
        tweet = json.loads(tweet)
        u_name = tweet['user']['screen_name']
        if u_name not in unames:
            unames.add(u_name)
            user = {'Uid':tweet['user']['id'],'Handle':u_name, 'Link': 'https://twitter.com/'+u_name}
            users.append(user)

users
```

```python
len(users)
```

```python
df = pd.DataFrame(users)
df
```

```python
df.to_csv('accounts.csv', index=False)
```
