```python
unames = set()
users = []

with open('tweets.jsonl', 'r') as json_lines:
    while True:
        tweet = json_lines.readline()
        if tweet == "":
            break
        tweet = json.loads(tweet)
        u_name = tweet['user']['screen_name']
        if u_name not in unames:
            unames.add(u_name)
            user = {'Uid':tweet['user']['id'],'Handle':u_name, 'Link': 'https://twitter.com/'+u_name}
            users.append(user)

users
```

```python
import h5py
with h5py.File('data/mini.h5','r') as f:
    all_words = [word.decode('utf-8') for word in f['mat']['axis1'][:]]
    all_embeddings = f['mat']['block0_values'][:]
```

```python
english_words = [word[6:] for word in all_words if word.startswith('/c/en/')]
len(english_words)
```

```python
english_word_indices = [i for i, word in enumerate(all_words) if word.startswith('/c/en/')]
english_embeddings = all_embeddings[english_word_indices]
index = {word: i for i, word in enumerate(english_words)}
```
